```python
import pandas as pd
import janitor
import sidetable
from IPython.display import display
```

```python
browser_group = {
    "ChromeBrowserPlugin": "Chrome",
    "Google Chrome": "Chrome",
    "Google": "Chrome",
    "FirefoxBrowserPlugin": "Firefox",
    "Firefox": "Firefox",
    "Microsoft Edge": "Microsoft",
    "MS Edge": "Microsoft",
    "SafariBrowserPlugin": "Safari",
    "Safari": "Safari",
    "Internet (Android Browser)": "Other Browsers",
    "Opera": "Other Browsers",
    "Opera Mini": "Other Browsers",
    "Internet Explorer": "Other Browsers",
    "Thunderbird": "Other Browsers"
}


df = (
    pd.read_csv("../data/yg/realityMine_web_2022-06-01_2022-06-30.csv")
    .assign(browser_group=lambda df_: df_["browser_vendor"].map(browser_group))
)
display(df.head())
df.info(verbose=True)
```

## Summ OS

### Visits level

```python
df.stb.freq(["os_name"], sort_cols=False)
```

```python
df.stb.freq(["os_name", "os_version"], sort_cols=False)
```

```python
df.stb.freq(["os_name", "os_version", "browser_vendor"], sort_cols=False)
```

### Ind level

```python
df_ind_os = (
    df
    .groupby(["caseid", "os_name"])
    .size()
    .reset_index(name="visits_count")
    .sort_values(["caseid", "visits_count"], ascending=(True, False))
    # Filter below 5 uses of an OS
    .query("visits_count > 5")
)
df_ind_os
```

```python
df_ind_os[df_ind_os.duplicated("caseid", keep=False)].reset_index(drop=True)
```

```python
df_ind_os.stb.freq(["os_name"], sort_cols=False)
```

## Browser

```python
df.stb.freq(["browser_vendor"], sort_cols=False)
```

### Ind level

```python
df_ind_browser = (
    df
    .groupby(["caseid", "browser_vendor"])
    .size()
    .reset_index(name="visits_count")
    .sort_values(["caseid", "visits_count"], ascending=(True, False))
    # Filter below 5 uses of a browser
    .query("visits_count > 5")    
)
df_ind_browser
```
