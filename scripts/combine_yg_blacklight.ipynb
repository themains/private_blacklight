```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
## YG data/Ignoring app/media play data
web_yg = pd.read_csv("../data/yg/realityMine_web_2022-06-01_2022-06-30.csv", low_memory = False)
dsk_yg = pd.read_csv("../data/yg/realityMine_web_desktop_2022-06-01_2022-06-30.csv")
mob_yg = pd.read_csv("../data/yg/realityMine_web_mobile_2022-06-01_2022-06-30.csv")
```

```python
dsk_yg.head()
```

```python
web_yg.head()
```

```python
web_yg.rename(columns={
    'page_duration': 'visit_duration',
    'start_time_utc': 'visit_time_utc'
}, inplace=True)
```

```python
web_yg['visit_time_local'] = pd.to_datetime(web_yg['date'] + ' ' + web_yg['time'])
web_yg['visit_time_local'].head()
```

```python
web_yg['page_views'] = 1
```

```python
web_yg = web_yg.assign(source = 'web')
dsk_yg = web_yg.assign(source = 'desktop_web')
mob_yg = mob_yg.assign(source = 'mob')

browsing = pd.concat([web_yg, dsk_yg, mob_yg], ignore_index=True)
browsing.head()
```

```python
profile = pd.read_csv("../data/yg/profile.csv")
profile.head()
```

```python
profile['caseid'] = pd.to_numeric(profile['caseid'], errors='coerce').astype('Int64')
```

```python
yg = profile.merge(
    browsing,
    left_on='caseid',
    right_on='caseid', 
    how='left'
)
yg.shape
```

```python
blacklight = pd.read_csv("../data/blacklight_domain.csv")
whotracksme = pd.read_csv("../data/whotracksme_domain.csv")
```

```python
blacklight.head()
```

```python
blacklight['filename'] = blacklight['filename'].str.replace('_', '.', regex=False)
```

```python
whotracksme.head()
```

```python
who_df = yg.merge(
    whotracksme,
    left_on = 'private_domain',
    right_on = 'domain_name', 
    how = 'left'
)
```

```python
final_df = who_df.merge(
    blacklight,
    left_on = 'private_domain',
    right_on = 'filename', 
    how = 'left'
)
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
columns_to_sum = [
    'ddg_join_ads', 'third_party_cookies', 'canvas_fingerprinting', 'session_recording', 
    'key_logging', 'fb_pixel', 'google_analytics', 'Trackers Per Page Load', 
    'Tracking Requests Per Page Load', 'Trackers Requests / All Requests', 
    'Data Saved', 'Advertising', 'Audio/Video Player', 'Customer Interaction', 'Hosting', 
    'Consent Management', 'Site Analytics', 'Misc', 'Utilities', 'Social Media', 'Adult Advertising'
]

grouped = final_df.groupby('caseid', as_index=False)[columns_to_sum].sum(min_count=1)
```

```python
grouped.fillna(0, inplace=True)
```

```python
grouped.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.99])
```

```python
grouped = final_df.groupby('caseid', as_index=False)[columns_to_sum].mean()
```
